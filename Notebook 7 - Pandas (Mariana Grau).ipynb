```python
#10- El máximo de datos faltantes

masnan = expvida.isnull().sum().max()
masnan
print ("El máximo de datos faltantes en una columna es", masnan)
```

```python
#10-Cantidad de filas totales, dividir el maximo valor de datos faltantes y multiplicarlo por 100
cantidad_total_filas = expvida.shape[0]

#Porcentaje de datos que representan los datos faltantes
porcentaje_de_datos = (cantidad_total_filas / masnan)*100

print(f"Máximo de datos faltantes en una columna: {masnan}")
print(f"Cantidad total de filas en expvida: {cantidad_total_filas}")
print(f"Porcentaje de datos faltantes en todo expvida: {porcentaje_de_datos:.2f}%")
```

```python
#11-Renombrar las columnas de expvida

expvida.rename(columns={'Life expectancy ': 'life_expectancy', ' BMI ': 'bmi', 'Measles ': 'measles'}, inplace=True)
expvida.columns

```

```python
#12-Promedio de la columna life expectancy 

promedio_life_expectancy = expvida['life_expectancy'].mean().round(2) #el round para que solo me muestre con dos decimales

print("Promedio de la columna Life Expectancy:", promedio_life_expectancy)

```

```python
#Promedio de life expectancy por país

promedio_life_expectancy_pais = expvida.groupby('Country')['life_expectancy'].mean().round(2)

print(promedio_life_expectancy_pais)
```

```python
#14-Expectativa de vida alta y baja

#Expectativa de vida alta - Pandas
expectativa_alta_Pandas =expvida ["life_expectancy"].max()

#Expectativa de vida alta - NumPy
expectativa_alta_Numpy = np.max (expvida['life_expectancy'])

#Expectativa de vida baja -Pandas
expectativa_baja_Pandas =expvida ["life_expectancy"].min()

#Expectativa de vida baja -NumPy
expectativa_baja_Numpy = np.min (expvida['life_expectancy'])

print("Expectativa de vida más alta es de:", expectativa_alta_Pandas, "(Pandas)")
print("Expectativa de vida más alta es de:", expectativa_alta_Numpy, "(Numpy)")
print("Expectativa de vida más baja es de:", expectativa_baja_Pandas, "(Pandas)")
print("Expectativa de vida más baja es de:", expectativa_baja_Numpy, "(Numpy)")
```

```python
#Filtro para ver las filas que tienen valor igual al maximo 

filas_maximo_Pandas = expvida[expvida['life_expectancy'] == expectativa_alta_Pandas]
filas_maximo_Numpy = expvida[expvida['life_expectancy'] == expectativa_alta_Numpy]

print("Los países que tienen una expectativa de vida alta son: Belgium, Finland, France, Germany," 
      "Italy,New Zealand, Norway, Portugal, Spain y Sweden.")
print(filas_maximo_Pandas)
print(filas_maximo_Numpy)

```

```python
#14-Filtro para ver las filas que tienen valor igual al minimo

filas_minimo_Pandas = expvida[expvida['life_expectancy'] == expectativa_baja_Pandas]
filas_minimo_Numpy = expvida[expvida['life_expectancy'] == expectativa_baja_Pandas]

print("El país que tiene la expectativa de vida más baja es Haiti:")
print(filas_minimo_Pandas)
print(filas_minimo_Numpy)
```

```python
#15-Categorias de la columna Status

status = expvida['Status'].nunique()

print("En la columna Status encontramos solo",status,"valores únicos.")
```

```python
#16-Paises dentro de las categorias de Status

paises_status = expvida['Status'].value_counts()

print("La cantidad de países que se encuentran dentro de cada categoria de Status son:", paises_status)
```

```python
#Países que presentaron algun caso de Sarampion (Measles) en el 2015

paises_sarampion_2015 = expvida[(expvida['measles'] > 0) & (expvida['Year'] == 2015)]
paises_sarampion = paises_sarampion_2015['measles'].count()
print(paises_sarampion, "países presentaron casos de Sarampión durante el 2015.")
```
